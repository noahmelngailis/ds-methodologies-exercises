>Write a function, months_to_years(tenure_months, df) that returns your dataframe with a new feature tenure_years, in complete years as a customer.

```python
def months_to_years(tenure_months, df):
    new_df = df
    new_df['tenure_years'] = (new_df.tenure/12).apply(np.ceil)
    return new_df
```

```python
tenure_months = df.tenure
new_df = months_to_years(tenure_months, df)
new_df
```

```python
categorical_var = new_df.tenure_years
continuous_var = new_df.monthly_charges
df = new_df
```

```python
def plot_categorical_and_continuous(categorical_var, continuous_var, df):
    fig, [ax1, ax2, ax3] = plt.subplots(ncols=3, nrows=1, figsize=(16, 9))
    
    l = sns.lineplot(x=categorical_var, y=continuous_var, data=df, ax=ax1)
    b = sns.boxplot(x=categorical_var, y=continuous_var, data=df, ax=ax2);
    v = sns.violinplot(x=categorical_var, y=continuous_var, data=df, ax=ax3);
    return fig, [ax1, ax2, ax3]
```
